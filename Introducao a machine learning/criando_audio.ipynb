```python
#Importando a biblioteca necessária para transformar texto em áudio.
from gtts import gTTS
```

```python
#Criando variáveis para os parâmetros necessários para criação do áudio.
say = """Bem-vindo ao primeiro teste, Bryan. E parabéns por estar se empenhando tanto."""
language = 'pt-br'

#Criando o objeto de áudio
gtts_object = gTTS(text = say,
                  lang = language,
                  slow = False)

#salvando o arquivo de áudio.
gtts_object.save("audio/gtts.wav")
```

```python
#Importando biblioteca utilizada para rodar o arquivo de áudio.
from IPython.display import Audio

#Reproduzindo arquivo criado.
Audio("audio/gtts.wav")
```
